```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import datetime
import time
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\Fran Zoia\\Documents\\Cursos\\Python\\python_data_science_big_data\\base_datos_2008.csv")
df2 = df[df["Origin"].isin(["ALT","HOU","IND"])]
df = df.head(500000)
```

```python
times = []
for i in np.arange(len(df)):
    times.append(datetime.datetime(year = 2008, month = df.loc[i, "Month"], day= df.loc[i,"DayofMonth"]))
```

```python
times[50000]
```

```python
df["Time"] = times
```

```python
data = df.groupby(by=["Time"], as_index=False)[["DepDelay", "ArrDelay"]].mean()
data.head()
```

```python
sns.lineplot(x="Time", y="DepDelay", data=data)
```
