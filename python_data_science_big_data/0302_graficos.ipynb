```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as ptl 

np.random.seed(0)
df = pd.read_csv("C:\\Users\\Fran Zoia\\Documents\\Cursos\\Python\\python_data_science_big_data\\base_datos_2008.csv", nrows=1000000)
df = df.sample(frac=1).head(100)
```

```python
ptl.scatter(x=df.DayofMonth, y=df.ArrDelay, s=df.Distance)
```

```python
ptl.scatter(x=df.DayofMonth, y=df.ArrDelay, s=df.Distance, alpha = .3, c = df .DayOfWeek.isin([6,7]))
ptl.title("Retraso en EEUU")
ptl.xlabel("Día del Mes")
ptl.ylabel("Retraso al Llegar")
ptl.ylim([0,150])
ptl.xticks([0,15,30])
ptl.text(x=28,y=120, s="Mi Vuelo")
ptl.show()
```
