```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as ptl 
import seaborn as sns

df = pd.read_csv("C:\\Users\\Fran Zoia\\Documents\\Cursos\\Python\\python_data_science_big_data\\base_datos_2008.csv", nrows=100000)
data = np.unique(df.DayOfWeek,return_counts = True)
labs = ["Lun","Mar","Mie", "Jue", "Vie", "Sab", "Dom"]
data
```

```python
ptl.pie(x = data[1], 
        #labels=labs,
        radius =.7,
        colors=sns.color_palette("hls",7),
        explode=(0,0,0,0.2,0,0,0.1),
        startangle = 90,
        autopct = "%1.1f%%")
ptl.legend(loc="upper left", labels = labs)
ptl.show()
```

```python
plt = sns.barplot(x = labs, y =data[1])
ptl,set()
```
